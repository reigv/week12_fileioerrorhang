## Intro 
func ptr is a var that store **memory addres of a function**
-- allow program to call a func **indirectly** through the pointer
-- useful for **dynamic dispatch** and **callback functions**

--> **Pointer to a function**


## Common uses of func ptr
1. **Callback functions**: func ptr can be used to pass a function as an argument to another function, allowing for dynamic behavior and flexibility in code design.

2. **Custom Comparison func**: for algo such as sorting, func ptr can be used to specify a custom comparison func, allowing for sorting based on different criteria.

3. **Table of func**: func ptr can be used to create a table of func, allowing for dynamic dispatch and polymorphism in object-oriented programming.


## Syntax of func ptr
```c
return_type (*pointer_name)(parameter_list);
```
- `return_type`: the data type of the value returned by the function.
- `pointer_name`: the name of the function pointer variable.    
- `parameter_list`: a comma-separated list of the data types of the parameters that the function takes. If the function takes no parameters, this can be left empty or specified as `void`.

## Example of func ptr
```c    
#include <stdio.h>
// Function that takes two integers and returns their sum
int add(int a, int b) {
    return a + b;
}   
int main() {
    // Declare a function pointer and assign it to the add function
    int (*func_ptr)(int, int) = add;
    
    // Use the function pointer to call the add function
    int result = func_ptr(5, 3);
    
    printf("The sum is: %d\n", result); // Output: The sum is: 8
    
    return 0;
}
```In this example, we declare a function `add` that takes two integers and returns their sum. We then declare a function pointer `func_ptr` that can point to a function with the same signature as `add`. We assign `func_ptr` to point to the `add` function and use it to call `add` indirectly, resulting in the output "The sum is: 8".

## why use func ptr?
1. **Flexibility** :: Dynamic behavior in code design, allowing for more modular and reusable code.

````python
void add() {std::cout << "Add\n";}
void subtract() {std::cout << "Subtract\n";}

void executeOperation(void (*operation)()) {
    operation(); // Call the function pointed to by operation
}

int main() {
    void (*operationPtr)(); // Declare a function pointer
    operationPtr = &add; // Point to the add function
    executeOperation(operationPtr); // Execute add

    operationPtr = &subtract; // Point to the subtract function
    executeOperation(operationPtr); // Execute subtract

    return 0;
}
````

2. **callback func**:
    very common in 
    - event-driven programming
    - library design
    - sorting
    

````python
#include <iostream>

void done() {
    std::cout << "Done!\n";
}

void process(void (*callback)()) {
    std::cout << "Processing...\n";
    callback(); // Call the callback function
}

int main() {
    process(done); // Pass the done function as a callback
    return 0;
}
````
